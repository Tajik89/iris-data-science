```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv('iris_data.csv')
```

```python
data_frame
```

```python
# Missing data
```

```python
for i in data_frame :
    data_frame[data_frame[i].notna()]
```

```python
# outlier data
```

```python
data_frame.boxplot()
plt.show()
```

```python
df_numeric = data_frame.select_dtypes(include=['float64', 'int64'])

Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)

IQR = Q3 - Q1

data_frame = data_frame[~((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
data_frame.boxplot()
plt.show()
```

```python
# Duplicate data
```

```python
data_frame = data_frame.drop_duplicates()
```

```python
data_frame.plot()
plt.show()
```
